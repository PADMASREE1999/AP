###Question 1:
Write a program in C to insert a new node at the beginning of a Singly Linked List.

```
#include<stdio.h>
#include<stdlib.h>
#include<conio.h>
struct node{
    int data;
    struct node *next;
}*s,*l,*c,*p;
//s - start, l - last, c - current, p - previous
void insertnode();
void deletenode();
void display();
int main()
{
    int x;
    s=l=c=p=NULL;
    while(1)
    {
        system("cls");//clrscr()
        printf("\n Single Linked List ");
        printf("\n 1. Insert ");
        printf("\n 2. Delete ");
        printf("\n 3. Display ");
        printf("\n 4. Exit");
        printf("\n Enter Your Choice: ");
        scanf("%d",&x);
        switch(x)
        {
            case 1:insertnode();
            break;
            case 2:deletenode();
            break;
            case 3:display();
            break;
            case 4:printf("\n Program is terminating...");
            return 0;
            break;
            
        }
    }
    return 0;
}
void insertnode()
{
    
    struct node *nn;
    nn=(struct node *)malloc(sizeof(struct node));
    printf("\n Enter the value : ");
    scanf("%d",&nn->data);
    nn->next=NULL;
    if(s==NULL)
    {
        s=l=nn;
    }
    else
    {
        nn->next=s;
        s=nn;
    }
    printf("\n Node is inserted at start");
    getch();
}
void deletenode()
{
    //delete node
}
void display()
{
    if(s==NULL)
    {
        printf("\n Single Linked List is empty");
        getch();
    }
    else
    {
        printf("\n nodes in listed list is : \n");
        for(c=s;c!=NULL;c=c->next)
        {
            printf("%d ",c->data);
            //printf("%d %u\n",C->Data,C->Next);
            
        }
        getch();
    }
    
}
```
